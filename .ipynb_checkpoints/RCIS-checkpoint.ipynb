```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (replace this with your actual data)
fitness = np.linspace(0.1, 1, 100)
precision = np.linspace(0.1, 1, 100)

# Create a grid of fitness and precision values
fitness, precision = np.meshgrid(fitness, precision)

# Calculate the harmonic mean
harmonic_mean = 2 * (fitness * precision) / (fitness + precision)

# Create a contour plot with colored lines and labels
plt.figure(figsize=(8, 6))
plt.xlim(0.3, 1.05)
plt.ylim(0.3, 1.05)

# Generate random points with values between 0 and 1
np.random.seed(42)  # For reproducibility
num_points = 20
num_par = 10
values_red = np.random.rand(num_points)  # Values for the red points
values_blue = np.random.rand(num_points)  # Values for the blue points

# Create a contour plot with colored lines and labels
contour = plt.contour(fitness, precision, harmonic_mean, levels=20, linestyles='dashed', cmap='viridis')
plt.clabel(contour, inline=True, fontsize=8, fmt='%1.2f', manual=False)


# Plot random points with red color and adjust darkness based on values
# scatter_red = plt.scatter(np.random.uniform(0.1, 1, num_points), np.random.uniform(0.1, 1, num_points),
#                           c=values_red, cmap='Reds', label='Random Points (Red)')
jitter = 0.003
scatter_red = plt.scatter(df_IMr_17['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMr_17['prc']/100+ np.random.uniform(-jitter, jitter, num_par),
                          c=df_IMr_17['par'], cmap='Reds', label='BPIC17 IMr', s=100, edgecolors='r', linewidth=0.5)

# Plot random points with blue color and adjust darkness based on values
scatter_blue = plt.scatter(df_IMf_17['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMf_17['prc']/100+ np.random.uniform(-jitter, jitter, num_par),
                          c=df_IMf_17['par'], cmap='Blues', label='BPIC17 IMf', s=100, edgecolors='b', linewidth=0.5)


# Plot points with diamond shapes
plt.scatter(df_IMr_18['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMr_18['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='D', c=df_IMr_18['par'], cmap='Reds', label='BPIC18 IMr', alpha=1, edgecolors='r', linewidth=0.5, s=100)
plt.scatter(df_IMf_18['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMf_18['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='D', c=df_IMf_18['par'], cmap='Blues', label='BPIC18 IMf', alpha=1, edgecolors='b', linewidth=0.5, s=100)


# Plot points with diamond shapes
plt.scatter(df_IMr_12['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMr_12['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='s', c=df_IMr_12['par'], cmap='Reds', label='BPIC12 IMr', alpha=1, edgecolors='r', linewidth=0.5, s=100)
plt.scatter(df_IMf_12['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMf_12['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='s', c=df_IMf_12['par'], cmap='Blues', label='BPIC12 IMf', alpha=1, edgecolors='b', linewidth=0.5, s=100)


# Plot points with diamond shapes
plt.scatter(df_IMr_hos['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMr_hos['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='^', c=df_IMr_hos['par'], cmap='Reds', label='Hospital IMr', alpha=1, edgecolors='r', linewidth=0.5, s=100)
plt.scatter(df_IMf_hos['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMf_hos['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='^', c=df_IMf_hos['par'], cmap='Blues', label='Hospital IMf', alpha=1, edgecolors='b', linewidth=0.5, s=100)


# Plot points with diamond shapes
plt.scatter(df_IMr_sep['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMr_sep['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='*', c=df_IMr_sep['par'], cmap='Reds', label='Sepsis IMr', alpha=1, edgecolors='r', linewidth=0.5, s=150)
plt.scatter(df_IMf_sep['fit']/100+ np.random.uniform(-jitter, jitter, num_par), df_IMf_sep['prc']/100+ np.random.uniform(-jitter, jitter, num_par), marker='*', c=df_IMf_sep['par'], cmap='Blues', label='Sepsis IMf', alpha=1, edgecolors='b', linewidth=0.5, s=150)





# Create colorbars for both sets of points
colorbar_red = plt.colorbar(scatter_red, ax=plt.gca(), label='sup in IMr', pad=-0.02,aspect=45)
colorbar_blue = plt.colorbar(scatter_blue, ax=plt.gca(), label='f in IMf', pad=0.01,aspect=45)



# Set axis labels, legend, and plot title
plt.xlabel('Fitness')
plt.ylabel('Precision')
plt.legend()
plt.title('comparison between IMr and IMf discovered models')

plt.savefig('compare_IMr_IMf.pdf', format='pdf')
# Show the plot
plt.show()
```
