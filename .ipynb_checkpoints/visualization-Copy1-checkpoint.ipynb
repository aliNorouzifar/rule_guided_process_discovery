```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import os
import json


figures_folder = r"E:\PADS\Projects\IMr_DKE\step_by_step\Experiments\figures"

config = {'c1':{'color':'blue','label':'c95s30','marker':'h','legend':'$R_1$'},
          'c2':{'color':"deepskyblue",'label':'c98s10','marker':'v','legend':'$R_2$'},
          'c3':{'color':"dodgerblue",'label':'c99s5','marker':'D','legend':'$R_3$'},
          'c4':{'color':"royalblue",'label':'c100s0','marker':'s','legend':'$R_4$'}}



all_options_labels = ['$R_1$', '$R_2$', '$R_3$','$R_4$']
all_options = ['c95s30', 'c98s10', 'c99s5','c100s0']

# config_list = ['c95s30','c98s10','c99s5','c100s0']


# logs = {"BPIC12":{'marker':'o'},
#         "BPIC17":{'marker':'h'}, 
#         "BPIC18":{'marker':'v'}, 
#         "RTFM":{'marker':'*'}, 
#         "Sepsis":{'marker':'D'},
#         "UWV":{'marker':'s'},
#         "HB":{'marker':'p'}}

# logs = {"BPIC12":{'marker':'o'},
#         "BPIC17":{'marker':'o'}, 
#         "BPIC18":{'marker':'o'}, 
#         "RTFM":{'marker':'o'}, 
#         "Sepsis":{'marker':'o'},
#         "UWV":{'marker':'o'},
#         "HB":{'marker':'o'}}
logs = {"BPIC12",
        "BPIC17", 
        "BPIC18", 
        "RTFM", 
        "Sepsis",
        "UWV",
        "HB"}
logs_list = ["BPIC12",
        "BPIC17", 
        "BPIC18", 
        "RTFM", 
        "Sepsis",
        "HB",
        "UWV"]

discoveries = {'IMr':{'label':'IMr','color':'blue'},
              'Imf':{'label':'IMf','color':'green'},
              'sm':{'label':'Split Miner','color':'orange'},
            'Imbi':{'label':'IMbi','color':'red'}
              }



```

```python

rules = {}
for ev in logs:
    rules[ev] = {op: set() for op in all_options}
    for op in all_options:
        path = f"E:\PADS\Projects\IMr_DKE\step_by_step\Experiments\declare_constraints\{op}\{ev}.json"
        with open(path, 'r') as file:
            report = json.load(file)
            for r in report["constraints"]:
                # print((r["template"],r["parameters"]))
                if len(r["parameters"])==1:
                    rules[ev][op].add((r["template"], r["parameters"][0][0]))
                else:
                    rules[ev][op].add((r["template"],r["parameters"][0][0],r["parameters"][1][0]))

comparison = {}
for ev in logs:
    comparison[ev] = {}
    for op1 in all_options:
        for op2 in all_options:
            comparison[ev][(op1,op2)] = len(rules[ev][op1]&rules[ev][op2])/len(rules[ev][op1]|rules[ev][op2])

print(comparison)
```

```python
# Number of rows and columns
num_logs = len(logs)
ncols = 4  # Two subplots per row
nrows = -(-num_logs // ncols)  # Ceiling division to determine rows

# Create subplots
fig, axes = plt.subplots(nrows=nrows, ncols=ncols, figsize=(10, 3 * nrows), constrained_layout=True)

# Flatten axes for easy indexing (works for both 1D and 2D layouts)
axes = axes.flatten()

# axis_labels = list(all_options)

for i, log in enumerate(logs_list):
    # Prepare data for each log
    matrix = pd.DataFrame(
        {var1: [comparison.get(log, {}).get((var1, var2), 0) for var2 in all_options] for var1 in all_options},
    )
    
    # Plot heatmap in each subplot
    sns.heatmap(matrix, annot=True, cmap='Blues', cbar=True, ax=axes[i])
    axes[i].set_title(f'rules similarity in {log}')
    axes[i].set_xticks([x + 0.5 for x in [0,1,2,3]])  # Set x-ticks positions
    axes[i].set_yticks([x + 0.5 for x in [0,1,2,3]])  # Set y-ticks positions
    axes[i].set_xticklabels(all_options_labels, rotation=0, ha='left')  # Set x-tick labels
    axes[i].set_yticklabels(all_options_labels, rotation=0)  # Set y-tick labels

# Hide any unused subplots
for j in range(len(axes)):
    if j >= num_logs:
        axes[j].axis('off')

fig.suptitle("Rules Similarity Across Event Logs", fontsize=16)

# Show the combined figure
plt.savefig(os.path.join(figures_folder,"rules_similarities.pdf"), bbox_inches='tight')
plt.show()
```

```python
# Process data for the bar chart
labels = list(runtime_all.keys())  # Event logs (e.g., 'BPIC12', 'BPIC17', ...)
configurations = list(runtime_all[labels[0]].keys())  # Configurations (e.g., 'c95s30', 'c98s10', ...)
num_configs = len(configurations)
x = np.arange(len(labels))  # X-axis positions for logs

colors = {'c95s30':{'color':'blue','legend':'IMr-$R_1$'},
          'c98s10':{'color':"deepskyblue",'legend':'IMr-$R_2$'},
          'c99s5':{'color':"dodgerblue",'legend':'IMr-$R_3$'},
          'c100s0':{'color':"royalblue",'legend':'IMr-$R_4$'},
         'IMr':{'legend':'IMr','color':'blue'},
              'Imf':{'legend':'IMf','color':'green'},
              'sm':{'legend':'Split Miner','color':'orange'},
            'Imbi':{'legend':'IMbi','color':'red'}}

# Bar width
bar_width = 0.1

# Plot
fig, ax = plt.subplots(figsize=(15, 8))

# Plot bars for each configuration
for i, config in enumerate(configurations):
    config_values = [runtime_all[log][config] for log in labels]  # Values for this config across logs
    print(colors[config]['color'])
    ax.bar(x + i * bar_width, config_values, bar_width, label=colors[config]['legend'],color=colors[config]['color'])

# Set y-axis to logarithmic scale
ax.set_yscale("log")

# Add labels, title, and legend
ax.set_xlabel("Event Logs", fontsize=14)
ax.set_ylabel("Runtime (seconds, log scale)", fontsize=14)
ax.set_title("Runtime Comparison Across Configurations and Event Logs", fontsize=16)
ax.set_xticks(x + bar_width * (num_configs - 1) / 2)
ax.set_xticklabels(labels, rotation=45, fontsize=12)
ax.legend(title="Configurations", fontsize=12)

fig.subplots_adjust(top=1) 
# Save and show the plot
plt.tight_layout()
plt.savefig(os.path.join(figures_folder,"runtime_comparison.pdf"), format="pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ast

# Event logs and options
logs = ["BPIC12", "BPIC17", "BPIC18", "RTFM", "Sepsis", "UWV", "HB"]
all_options = ['c95s30', 'c99s5', 'c98s10', 'c100s0']

# Generate dummy data (replace this with your actual data)
np.random.seed(42)
support_values = np.linspace(0.1, 1, 9)  # Support parameter values
data = {}
max_y = {}
for log in logs:
    data[log] = {}
    max_y[log] = {}
    file_path_IMbi = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMbi_{log}.csv"
    file_path_IMr = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMr_{log}.csv"
    df_IMbi = pd.read_csv(file_path_IMbi, index_col=0)
    df_IMr = pd.read_csv(file_path_IMr, index_col=0)
    for option in all_options:
        data[log][option] = {
            'IMbi': [ast.literal_eval(x)['support_cost'] for x in df_IMbi[option].tolist()],
            'IMr': [ast.literal_eval(x)['support_cost'] for x in df_IMr[option].tolist()]
        }
        max_y[log][option] ={'IMbi':ast.literal_eval(df_IMbi[option].tolist()[0])['N.rules'],
                'IMr':ast.literal_eval(df_IMr[option].tolist()[0])['N.rules']}

# Create a big figure with rows = logs and columns = options
num_logs = len(logs)
num_options = len(all_options)
fig, axs = plt.subplots(num_logs, num_options, figsize=(16, 12), sharex=True)

# Plot each subplot
for i, log in enumerate(logs):
    for j, option in enumerate(all_options):
        ax = axs[i, j]
        
        # Get data for this log and option
        imbi_deviations = data[log][option]['IMbi']
        imr_deviations = data[log][option]['IMr']
        
        # Scatter plot for IMbi (red) and IMr (blue)
        ax.scatter(support_values, imbi_deviations, color='red', label='IMbi', zorder=2)
        ax.scatter(support_values, imr_deviations, color='blue', label='IMr', zorder=2)
        
        # Line plot connecting the points
        ax.plot(support_values, imbi_deviations, color='red', linestyle='-', linewidth=1, zorder=1)
        ax.plot(support_values, imr_deviations, color='blue', linestyle='-', linewidth=1, zorder=1)
        
        # Set dynamic y-axis limits
        max_value = max(max(imbi_deviations), max(imr_deviations))
        # max_value = max_y[log][option]['IMbi']
        ax.set_ylim(0, max_value * 1.1)  # Add a 10% margin for better visualization
        
        # Add title and labels
        if i == 0:
            ax.set_title(option, fontsize=10)  # Title for each column (option)
        if j == 0:
            ax.set_ylabel(log, fontsize=10)  # Label for each row (event log)
        
        ax.set_xlabel("Support", fontsize=8)
        ax.set_ylabel("Deviations", fontsize=8)

        # Customize tick labels
        ax.tick_params(axis='both', which='major', labelsize=7)

# Add a legend to the first subplot
axs[0, 0].legend(loc='upper left', fontsize=8)

# Add a main title
fig.suptitle("Number of Deviations vs Support for Different Event Logs and Options", fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save and show the plot
plt.savefig("big_figure_deviations.pdf", format="pdf")
plt.show()
```

```python
import pandas as pd

# Load the lookup table (replace with your table file)
lookup_table = pd.read_csv("E:\PADS\Projects\IMr_DKE\lookup_table.csv", index_col=0)

# Define the mapping
mapping = {
    "seq": "\\textcolor{blue}{*}",
    "par": "\\textcolor{red}{*}",
    "xor": "\\textcolor{green}{*}",
    "loop": "\\textcolor{orange}{*}",
}

# Convert the table entries
latex_rows = []
for index, row in lookup_table.iterrows():
    latex_row = []
    for entry in row:
        if pd.isna(entry):
            latex_row.append("")  # Leave empty for NaN
        else:
            # Process the set and replace valid values
            processed = " ".join(mapping.get(value, "") for value in eval(entry) if value in mapping)
            latex_row.append(processed)
    latex_rows.append(" & ".join(latex_row) + " \\\\")

# Generate the LaTeX code
latex_code = """
\\begin{table}[ht]
\\centering
\\begin{tabular}{""" + " | ".join(["c"] * len(lookup_table.columns)) + """}
\\hline
""" + " & ".join(lookup_table.columns) + """ \\\\ \\hline
""" + "\n".join(latex_rows) + """
\\hline
\\end{tabular}
\\caption{Lookup Table}
\\label{tab:lookup_table}
\\end{table}
"""

# Print the LaTeX code
print(latex_code)
```
