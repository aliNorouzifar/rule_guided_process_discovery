```python
import pandas as pd
import ast
import matplotlib.pyplot as plt
import numpy as np


config = {'c1':{'color':'blue','label':'c95s30','marker':'h'},
          'c2':{'color':"deepskyblue",'label':'c98s10','marker':'v'},
          'c3':{'color':"dodgerblue",'label':'c99s5','marker':'D'},
          'c4':{'color':"royalblue",'label':'c100s0','marker':'s'}}


# logs = {"BPIC12":{'marker':'o'},
#         "BPIC17":{'marker':'h'}, 
#         "BPIC18":{'marker':'v'}, 
#         "RTFM":{'marker':'*'}, 
#         "Sepsis":{'marker':'D'},
#         "UWV":{'marker':'s'},
#         "HB":{'marker':'p'}}

# logs = {"BPIC12":{'marker':'o'},
#         "BPIC17":{'marker':'o'}, 
#         "BPIC18":{'marker':'o'}, 
#         "RTFM":{'marker':'o'}, 
#         "Sepsis":{'marker':'o'},
#         "UWV":{'marker':'o'},
#         "HB":{'marker':'o'}}
logs = {"BPIC12",
        "BPIC17", 
        "BPIC18", 
        "RTFM", 
        "Sepsis",
        "UWV",
        "HB"}

discoveries = {'IMr':{'label':'IMr','color':'blue'},
              'Imf':{'label':'IMf','color':'green'},
              'sm':{'label':'Split Miner','color':'orange'},
            'Imbi':{'label':'IMbi','color':'red'}
              }

```

```python
# read IMr statistics (fitness, precision)
stats_IMr = {}
for exp in logs:
    stats_IMr[exp] = {}
    for cf in {'c1','c2','c3','c4'}:
        df = pd.read_csv(f"E:\PADS\Projects\IMr_DKE\step_by_step\Experiments\Stats_rules\stats_{exp}.csv", sep=",")
        stats_IMr[exp][cf] = df[df['config']==cf]

# read IMbi statistics (fitness, precision)
stats_IMbi = {}
for exp in logs:
    stats_IMbi[exp] = {}
    df = pd.read_csv(f"E:\PADS\Projects\IMr_DKE\step_by_step\Experiments\Stats_rules\stats_{exp}_models.csv", sep=",")
    stats_IMbi[exp] = df[df['algorithm']=='Imbi']
```

```python


fitness = np.linspace(0.1, 1, 100)
precision = np.linspace(0.1, 1, 100)
fitness, precision = np.meshgrid(fitness, precision)
harmonic_mean = 2 * (fitness * precision) / (fitness + precision)

# Create subplots in a 4x2 layout
num_logs = len(logs)
fig, axs = plt.subplots(4, 2, figsize=(9, 16), sharex=True, sharey=True)
axs = axs.flatten()  # Flatten the 4x2 grid into a 1D array for easy indexing

jitter = 0.01  # For jitter in scatter plot

for idx, (l, ax) in enumerate(zip(logs, axs)):
    ax.set_xlim(0.4, 1.05)
    ax.set_ylim(0.4, 1.05)
    
    # Contour for harmonic mean
    contour = ax.contour(fitness, precision, harmonic_mean, levels=20, linestyles='dashed', cmap='viridis')
    ax.clabel(contour, inline=True, fontsize=8, fmt='%1.2f', manual=False)
    
    # Scatter plots for each configuration
    for cf in {'c1', 'c2', 'c3', 'c4'}:
        f_data = stats_IMr[l][cf]["Trace Fitness Positive"]
        p_data = stats_IMr[l][cf]["Precision"]
        ax.scatter(
            f_data + np.random.uniform(-jitter, jitter, len(f_data)),
            p_data + np.random.uniform(-jitter, jitter, len(p_data)),
            color=config[cf]['color'],
            label=config[cf]['label'],
            s=100,
            linewidth=0.5,
            marker=config[cf]['marker']
        )

    f_data = stats_IMbi[l]["Trace Fitness Positive"]
    p_data = stats_IMbi[l]["Precision"]
    ax.scatter(
        f_data + np.random.uniform(-jitter, jitter, len(f_data)),
        p_data + np.random.uniform(-jitter, jitter, len(p_data)),
        color=discoveries['Imbi']['color'],
        label=discoveries['Imbi']['label'],
        s=100,
        linewidth=0.5,
        marker='o'
    )

    # Set titles and labels for subplots
    ax.set_title(f'Log: {l}')
    ax.set_xlabel('Fitness')
    ax.set_ylabel('Precision')
    ax.legend()

# Remove any unused subplots if there are fewer than 8 logs
for ax in axs[num_logs:]:
    fig.delaxes(ax)

# Overall title for the figure
fig.suptitle('Comparison between IMr and IMf Discovered Models', fontsize=16)

# Save and show the plot
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig('compare_IMr_IMf_subplots.pdf', format='pdf')
plt.show()
```

```python

# Event logs and options
# logs = ["BPIC12", "BPIC17", "BPIC18", "RTFM", "Sepsis", "UWV", "HB"]
all_options = ['c95s30', 'c99s5', 'c98s10', 'c100s0']

# Generate dummy data (replace this with your actual data)
np.random.seed(42)
support_values = np.linspace(0.1, 1, 9)  # Support parameter values
data = {}
max_y = {}
for log in logs:
    data[log] = {}
    max_y[log] = {}
    file_path_IMbi = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMbi_{log}.csv"
    file_path_IMr = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMr_{log}.csv"
    df_IMbi = pd.read_csv(file_path_IMbi, index_col=0)
    df_IMr = pd.read_csv(file_path_IMr, index_col=0)
    for option in all_options:
        data[log][option] = {
            'IMbi': [ast.literal_eval(x)['support_cost'] for x in df_IMbi[option].tolist()],
            'IMr': [ast.literal_eval(x)['support_cost'] for x in df_IMr[option].tolist()]
        }
        max_y[log][option] ={'IMbi':ast.literal_eval(df_IMbi[option].tolist()[0])['N.rules'],
                'IMr':ast.literal_eval(df_IMr[option].tolist()[0])['N.rules']}

# Create a big figure with rows = logs and columns = options
num_logs = len(logs)
num_options = len(all_options)
fig, axs = plt.subplots(num_logs, num_options, figsize=(16, 12), sharex=True)

# Plot each subplot
for i, log in enumerate(logs):
    for j, option in enumerate(all_options):
        ax = axs[i, j]
        
        # Get data for this log and option
        imbi_deviations = data[log][option]['IMbi']
        imr_deviations = data[log][option]['IMr']
        
        # Scatter plot for IMbi (red) and IMr (blue)
        ax.scatter(support_values, imbi_deviations, color='red', label='IMbi', zorder=2)
        ax.scatter(support_values, imr_deviations, color='blue', label='IMr', zorder=2)
        
        # Line plot connecting the points
        ax.plot(support_values, imbi_deviations, color='red', linestyle='-', linewidth=1, zorder=1)
        ax.plot(support_values, imr_deviations, color='blue', linestyle='-', linewidth=1, zorder=1)
        
        # Set dynamic y-axis limits
        max_value = max(max(imbi_deviations), max(imr_deviations))
        # max_value = max_y[log][option]['IMbi']
        ax.set_ylim(0, max_value * 1.1)  # Add a 10% margin for better visualization
        
        # Add title and labels
        if i == 0:
            ax.set_title(option, fontsize=10)  # Title for each column (option)
        if j == 0:
            ax.set_ylabel(log, fontsize=10)  # Label for each row (event log)
        
        ax.set_xlabel("Support", fontsize=8)
        ax.set_ylabel("Deviations", fontsize=8)

        # Customize tick labels
        ax.tick_params(axis='both', which='major', labelsize=7)

# Add a legend to the first subplot
axs[0, 0].legend(loc='upper left', fontsize=8)

# Add a main title
fig.suptitle("Number of Deviations vs Support for Different Event Logs and Options", fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save and show the plot
plt.savefig("big_figure_deviations.pdf", format="pdf")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import ast

# Event logs and options
logs = ["BPIC12", "BPIC17", "BPIC18", "RTFM", "Sepsis", "UWV", "HB"]
all_options = ['c95s30', 'c99s5', 'c98s10', 'c100s0']

# Generate dummy data (replace this with your actual data)
np.random.seed(42)
support_values = np.linspace(0.1, 1, 9)  # Support parameter values
data = {}
max_y = {}
for log in logs:
    data[log] = {}
    max_y[log] = {}
    file_path_IMbi = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMbi_{log}.csv"
    file_path_IMr = f"E:\\PADS\\Projects\\IMr_DKE\\step_by_step\\Experiments\\Stats_rules\\stats_IMr_{log}.csv"
    df_IMbi = pd.read_csv(file_path_IMbi, index_col=0)
    df_IMr = pd.read_csv(file_path_IMr, index_col=0)
    for option in all_options:
        data[log][option] = {
            'IMbi': [ast.literal_eval(x)['support_cost'] for x in df_IMbi[option].tolist()],
            'IMr': [ast.literal_eval(x)['support_cost'] for x in df_IMr[option].tolist()]
        }
        max_y[log][option] ={'IMbi':ast.literal_eval(df_IMbi[option].tolist()[0])['N.rules'],
                'IMr':ast.literal_eval(df_IMr[option].tolist()[0])['N.rules']}

# Create a big figure with rows = logs and columns = options
num_logs = len(logs)
num_options = len(all_options)
fig, axs = plt.subplots(num_logs, num_options, figsize=(16, 12), sharex=True)

# Plot each subplot
for i, log in enumerate(logs):
    for j, option in enumerate(all_options):
        ax = axs[i, j]
        
        # Get data for this log and option
        imbi_deviations = data[log][option]['IMbi']
        imr_deviations = data[log][option]['IMr']
        
        # Scatter plot for IMbi (red) and IMr (blue)
        ax.scatter(support_values, imbi_deviations, color='red', label='IMbi', zorder=2)
        ax.scatter(support_values, imr_deviations, color='blue', label='IMr', zorder=2)
        
        # Line plot connecting the points
        ax.plot(support_values, imbi_deviations, color='red', linestyle='-', linewidth=1, zorder=1)
        ax.plot(support_values, imr_deviations, color='blue', linestyle='-', linewidth=1, zorder=1)
        
        # Set dynamic y-axis limits
        max_value = max(max(imbi_deviations), max(imr_deviations))
        # max_value = max_y[log][option]['IMbi']
        ax.set_ylim(0, max_value * 1.1)  # Add a 10% margin for better visualization
        
        # Add title and labels
        if i == 0:
            ax.set_title(option, fontsize=10)  # Title for each column (option)
        if j == 0:
            ax.set_ylabel(log, fontsize=10)  # Label for each row (event log)
        
        ax.set_xlabel("Support", fontsize=8)
        ax.set_ylabel("Deviations", fontsize=8)

        # Customize tick labels
        ax.tick_params(axis='both', which='major', labelsize=7)

# Add a legend to the first subplot
axs[0, 0].legend(loc='upper left', fontsize=8)

# Add a main title
fig.suptitle("Number of Deviations vs Support for Different Event Logs and Options", fontsize=16)

# Adjust layout
plt.tight_layout(rect=[0, 0, 1, 0.95])

# Save and show the plot
plt.savefig("big_figure_deviations.pdf", format="pdf")
plt.show()
```
